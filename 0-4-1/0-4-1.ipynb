```python
import numpy as np
import re
import string

labels = ["byr", "iyr", "eyr", "hgt", "hcl", "ecl", "pid"]

file = open('C:/Users/Niels/source/repos/aoc-2020/0-4-1/input.txt').readlines()

keys = []
values = []

t = ''
v = ''
for line in file:
    # handle splits between keys
    if line == '\n':
        keys.append(t)
        values.append(v)
        t = ''
        v = ''
        continue
    for el in line.split():
        spl = el.split(':')
        key = spl[0]
        if key != "cid": # cid is optional
            t += key + " "
            v += spl[1] + " "

print(keys[0])

for i in range(len(keys)):
    keys[i] = keys[i].strip()
    values[i] = values[i].strip()


def vbyr(byr):
    return int(byr) >= 1920 and int(byr) <= 2002

def viyr(iyr):
    return int(iyr) >= 2010 and int(iyr) <= 2020

def veyr(eyr):
    return int(eyr) >= 2020 and int(eyr) <= 2030

def vhgt(hgt):
    if str(hgt).endswith('in'):
        return int(hgt[:-2]) >= 59 and int(hgt[:-2]) <= 76    
    elif str(hgt).endswith('cm'):
        return int(hgt[:-2]) >= 150 and int(hgt[:-2]) <= 193
    else:
        return False

def vhcl(hcl):
    return hcl[0] == '#' and is_hex(hcl[1:])

def is_hex(s):
     hex_digits = set(string.hexdigits)
     # if s is long, then it is faster to check against a set
     return all(c in hex_digits for c in s)

print('testing vhcl:', vhcl('#123abc'), vhcl('#123abz'))

hair_colours = ['amb', 'blu', 'brn', 'gry', 'grn', 'hzl', 'oth']

def vecl(ecl):
    return ecl in hair_colours
print('testing vecl:', vecl('amb'), vecl('brn'))


def vpid(pid):
    return len(pid) == 9 and pid.isdigit()
print('testing vpid:', vpid('000000001'), vpid('0123456789'))

    
rule_dict = {
    'byr': vbyr,
    'iyr': viyr,
    'eyr': veyr,
    'hgt': vhgt,
    'hcl': vhcl,
    'ecl': vecl,
    'pid': vpid,
}


def checkValid(record, value):
    record_spl = record.split()
    value_spl = value.split()
    if len(record_spl) == len(labels) and all(rule_dict[key](value) for key, value in zip(record_spl, value_spl)):
    # all(key in labels for key in record_spl) and
        return True
    return False

valid = [checkValid(key, value) for key, value in zip(keys, values)]
print(valid.count(True))
```
